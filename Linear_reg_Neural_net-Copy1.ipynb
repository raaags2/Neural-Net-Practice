```python
import os
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam, SGD
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
os.chdir(r'F:\backup\practice\python\Machine_learning\Part 2 - Regression\Section 4 - Simple Linear Regression\Simple_Linear_Regression')
```

```python
dt = pd.read_csv('Salary_Data.csv')
```

```python
dt.describe()
```

```python
model = Sequential()
```

```python
x = dt['YearsExperience']
y = dt['Salary']
```

```python
model.add(Dense(1,activation='linear',input_shape=(1,)))
```

```python
model.summary()
```

```python
model.compile(Adam(lr=0.8),'mean_squared_error')
```

```python
model.fit(x,y,epochs=400)
```

```python
yp = model.predict(x)
```

```python
dt.plot(x='YearsExperience',y='Salary',kind='scatter')
plt.scatter(x,yp,color='red')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y,yp)
```

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
x1 = x.reshape(-1,1)
lreg.fit(x1,y)
ypl = lreg.predict(x1)
r2_score(y,ypl)
```

```python
dt.plot(x='YearsExperience',y='Salary',kind='scatter')
plt.scatter(x,ypl,color='red')
plt.show()
```
